```python
import pandas as pd
df = pd.read_csv("shirts.csv")
print(df)
```

## kNN Example w/ Sci-kit learn

When creating a model in Sci-kit learn, it needs two things:

1. $X$: feature matrix (2D data structure)
    - rows are instances and columns are attributes/features
1. $y$: class column (1D data structure)
    - the attribute/feature we want to predict
    - AKA: target array, target vector, labels, etc

*_train* and *_test* are used denote the training and testing sets, respectively

```python
X_train = df.drop(["t-shirt size"], axis="columns")
print(X_train)
```

```python
y_train = df["t-shirt size"]
print(y_train)
```

```python
X_unseen = [[161, 63]]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
print(X_train_scaled)
```

```python
X_unseen_scaled = scaler.transform(X_unseen)
print(X_unseen_scaled)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier(n_neighbors = 2, metric = "euclidean")

knn_clf.fit(X_train_scaled, y_train)

y_predicted = knn_clf.predict(X_unseen_scaled)

print(y_predicted)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()

tree_clf.fit(X_train_scaled, y_train)

y_predicted = tree_clf.predict(X_unseen_scaled)
print(y_predicted)
```
