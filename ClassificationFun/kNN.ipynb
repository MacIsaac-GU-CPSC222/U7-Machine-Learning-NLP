```python
import pandas as pd

df = pd.read_csv("../shirts.csv")
print(df)
```

## kNN Example w/ Sci-kit Learn

When creating a model in Sci-kit learn, it needs two things:
1. $X$: feature matrix (2D data structure)
    - rows are instances and columns are attributes/features
    - does NOT include feature we are trying to predict
1. $y$: class column (1D data structure)
    - the attribute/feature we want to predict
    - AKA: target array, target vector, labels, etc

*_train* and *_test* are used to denote the training and testing sets, respectively

```python
X_train = df.drop(["t-shirt size"], axis="columns")
print(X_train)
```

```python
y_train = df["t-shirt size"]
y_train
```

```python
X_unseen = [[161, 63], [172, 60]]
```

```python
# let's scale/normalize our feature values!
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train) # fitting it will determine the minimum and maximum values for each feature

X_train_scaled = scaler.transform(X_train)
print(X_train_scaled)
```

```python
X_unknown_scaled = scaler.transform(X_unseen)
print(X_unknown_scaled)
```

### Using kNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier(n_neighbors=3, metric="euclidean")

# builds/trains
knn_clf.fit(X_train_scaled, y_train)

# predict
y_predicted = knn_clf.predict(X_unknown_scaled)
print(y_predicted)
```

## Closing Thoughts on kNN

- Inefficient algorithm, but it is a great first alg because it is easy to understand and implement
- What if you have features that are categorical?
    - Encode values as integers (0, 1, 2, ...)
        - Sci-kit learn's LabelEncoder can help with this
        - Ok if category is ordinal (can be ordered)
        - Numbering adds an inherent ordering that will effect distances. For categories that are nominal, ordering could effect predictions
    - Use another distance function or create your own!
- kNN is one of MANY supervised ML algs
    - Decision trees
    - Random forests
    - Naive Bayes
    - Support vector machines
    - Neural Networks

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()

# builds/trains
tree_clf.fit(X_train_scaled, y_train)

# predict
y_predicted = tree_clf.predict(X_unknown_scaled)
print(y_predicted)
```
